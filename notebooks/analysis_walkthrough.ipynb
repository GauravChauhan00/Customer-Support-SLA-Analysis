Beginner-friendly analysis walkthrough. Dataset synthetic hai.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
df = pd.read_csv(ROOT / 'data/processed/support_tickets_clean.csv')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False).head(10)
```

```python
df['status'].value_counts()
```

```python
resolved = df[df['status'] == 'Resolved'].copy()
round(resolved['overall_sla_met'].mean() * 100, 2)
```

```python
df.groupby('assigned_team')['ticket_id'].count().sort_values(ascending=False)
```

```python
resolved.groupby('assigned_team')['overall_sla_met'].mean().mul(100).round(2).sort_values()
```

```python
df.groupby('category')['ticket_id'].count().sort_values(ascending=False).head(10).plot(kind='barh', title='Top Ticket Categories');
plt.tight_layout();
```
